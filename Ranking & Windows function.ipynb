```
#ranking Method 1

df = spark.read.csv("/FileStore/tables/walmart-3.csv",header=True,inferSchema=True)
df.show()
```

```
df = df.orderBy(df['Weekly_Sales'].desc())
df.show()
```

```
from pyspark.sql import functions as f

df = df.withColumn('ranky',f.monotonically_increasing_id())

df.show()
```

```
df = df.orderBy('Temperature')
df.show()
```

```
df.filter((df['ranky']>=100)&(df['ranky']<=200)).show()
```

```
#**************************************************************************
# ranking row_number

ff = spark.read.csv("/FileStore/tables/walmart-3.csv",header=True,inferSchema=True)
ff.show()
```

```
# ranking rownumber

from pyspark.sql import functions as f

from pyspark.sql.window import Window
rule = Window.orderBy(f.col('Weekly_Sales').desc())

ff = ff.withColumn('ranky',f.row_number().over(rule))
ff.show()

#**************************************************************************
```

```
#**************************************************************************
# ranking row_number with Different dataset


ff = spark.read.csv("/FileStore/tables/pokemon.csv",header=True,inferSchema=True)
ff.show()
```

```
from pyspark.sql import functions as f

from pyspark.sql.window import Window
rule = Window.orderBy(f.col('Speed').desc())

ff = ff.withColumn('ranky',f.row_number().over(rule))
ff.show()

#**************************************************************************
```

```
#**************************************************************************
# ranking with rank method

ff = spark.read.csv("/FileStore/tables/pokemon.csv",header=True,inferSchema=True)

from pyspark.sql.window import Window
from pyspark.sql import functions as f

rule = Window.orderBy(f.col('Speed').desc())

ff = ff.withColumn('rank',f.rank().over(rule))
ff.show()

#**************************************************************************
```

```
#**************************************************************************

# ranking with dense_rank method

ff = spark.read.csv("/FileStore/tables/pokemon.csv",header=True,inferSchema=True)

from pyspark.sql.window import Window
from pyspark.sql import functions as f

rule = Window.orderBy(f.col('Speed').desc())

ff = ff.withColumn('rank',f.dense_rank().over(rule))
ff.show()

#**************************************************************************
```

```
#**************************************************************************

#department wise ranking

ff = spark.read.csv("/FileStore/tables/pokemon.csv",header=True,inferSchema=True)
ff.show()
```

```
rule = Window.partitionBy('Type 1').orderBy(f.col('Speed').desc())

ff = ff.withColumn('rank',f.row_number().over(rule))
ff.show()
```
